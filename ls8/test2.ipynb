```python
#calculate values of array
numbers=[10,20,30,40,50]
s=0
for n in numbers:
    s=s+n
print('Sum of list items:',s)
```

```python
#calculate values of array use sum()
numbers=[10,20,30,40,50]
print('Sum of list items:',sum(numbers))
```
