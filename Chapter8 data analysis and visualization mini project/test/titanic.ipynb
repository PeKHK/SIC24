```python
import pandas as pd
import matplotlib.pyplot as plt
titanic=pd.read_csv("C:/Users/Windows 10/Downloads/Titanic.csv")
```

```python
titanic.info()
```

```python
#1.
passenger_13=titanic.iloc[12]
passenger_666=titanic.iloc[665]
print(f'Passenger 13:\n{passenger_13}')
print()
print(f'Passenger 666:\n{passenger_666}')
```

```python
#2.
total_Passenger=len(titanic)
surviver_count=titanic['Survived'].sum()
surviver_rate=surviver_count/total_Passenger*100
print(f'surviver_rate={surviver_rate}')
```

```python
#3.
max_age=titanic['Age'].max()
oldest_Passenger=titanic[titanic['Age']==max_age]
print(f'Oldest Passenger Is:')
print(oldest_Passenger)
```

```python
#4.
totalfare=titanic.groupby('Sex')['Fare'].sum()
print(f'Totalfare={totalfare}')
if totalfare['male']>totalfare['female']:
    print('Male payed more than female')
else:
    print('Female payed more than male')
```

```python
#5.
children=titanic[titanic['Age']<18]
adults=titanic[titanic['Age']>=18]

children_rate=children['Survived'].mean()
adults_rate=adults['Survived'].mean()

print(f'Children survived rate={children_rate}')
print(f'Adults survived rate={adults_rate}')

if children_rate>adults_rate:
    print('Children has survived more than adults')
else:
    print('Adults has survived more than children')
```

```python
#6.
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.hist(titanic['Age'].dropna(),bins=20,color='green',edgecolor='black')
plt.title('Age Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.subplot(1,2,2)
plt.hist(titanic['Fare'].dropna(),bins=20,color='yellow',edgecolor='black')
plt.title('Fare Histogram')
plt.xlabel('Fare')
plt.ylabel('Frequency')

plt.show()
```

```python
#7.
male=titanic[titanic['Sex']=='male']
female=titanic[titanic['Sex']=='female']

plt.figure(figsize=(15,5))

plt.scatter(male['Age'],male['Fare'],color='skyblue',label='Male',alpha=0.5)
plt.scatter(female['Age'],female['Fare'],color='red',label='Female',alpha=0.5)


plt.ylabel('Fare')

plt.legend()
plt.show()
```
